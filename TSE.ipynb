```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
pd.set_option('max_columns', None)
# É necessário deixar registado que o arquivo original tinha 4GB, meu computador não conseguiu nem abrir o arquivo, por isso
# utilizei o aplicativo Split para dividir o arquivo em vários menores, irei realizar o tratamento em um desses arquivos, que
# ficou com 906MB, por questão de limitação computacional não será possível utilizar todo o arquivo, mas como essa amostra contém
# quase 1/4 dos dados e 2 milhões de linhas ela deve representar bem os problemas presentes no conjunto
df = pd.read_csv('C:/Users/caiom/IA/Arquivos/TCC/BD/TSE/votacao_candidato_munzona_2022_BRASIL_0.csv', encoding='Latin 1',sep=";")
df.head()
```

```python
# Vale pontuar que apesar das muitas abreviações este aquivo veio bem documentado, com um arquivo pdf anexado explicando cada 
# coluna

# Iremos agora remover as colunas irrelevantes
df.drop(['DT_GERACAO'],axis=1,inplace=True)
df.drop(['HH_GERACAO'],axis=1,inplace=True) 
df.drop(['ANO_ELEICAO'],axis=1,inplace=True)
df.drop(['CD_TIPO_ELEICAO'],axis=1,inplace=True)
df.drop(['NM_TIPO_ELEICAO'],axis=1,inplace=True)
df.drop(['CD_ELEICAO'],axis=1,inplace=True)
df.drop(['DS_ELEICAO'],axis=1,inplace=True)
df.drop(['DT_ELEICAO'],axis=1,inplace=True)
df.drop(['NM_UE'],axis=1,inplace=True)
df.drop(['CD_MUNICIPIO'],axis=1,inplace=True)
df.drop(['CD_CARGO'],axis=1,inplace=True)
df.drop(['NR_CANDIDATO'],axis=1,inplace=True)
df.drop(['NM_CANDIDATO'],axis=1,inplace=True)
df.drop(['NM_URNA_CANDIDATO'],axis=1,inplace=True)
df.drop(['NM_SOCIAL_CANDIDATO'],axis=1,inplace=True)
df.drop(['CD_SITUACAO_CANDIDATURA'],axis=1,inplace=True)
df.drop(['CD_DETALHE_SITUACAO_CAND'],axis=1,inplace=True)
df.drop(['NM_PARTIDO'],axis=1,inplace=True)
df.drop(['NR_FEDERACAO'],axis=1,inplace=True)
df.drop(['NM_FEDERACAO'],axis=1,inplace=True)
df.drop(['SG_FEDERACAO'],axis=1,inplace=True)
df.drop(['SQ_COLIGACAO'],axis=1,inplace=True)
df.drop(['NM_COLIGACAO'],axis=1,inplace=True)
df.drop(['DS_COMPOSICAO_COLIGACAO'],axis=1,inplace=True)
df.drop(['CD_SIT_TOT_TURNO'],axis=1,inplace=True)
df.drop(['DS_SIT_TOT_TURNO'],axis=1,inplace=True)

# 26 colunas foram classificadas como irrelevantes, restando apenas 18 colunas para análise
df.head()
```

```python
# Agora adequaremos os tipos de dados
df.dtypes
```

```python
df['SQ_CANDIDATO']=df['SQ_CANDIDATO'].astype(str)
df['NR_PARTIDO']=df['NR_PARTIDO'].astype(str)
# Apesar de SQ_CANDIDATO e NR_PARTIDO serem de fato valores numéricos não há sentido em realizar operaçõe matemáticas com eles,
# portanto os deixaremos como String, como essa mudança é opcional e não afeta a utilização dos dados, a mudança será ignorada
# no relatório
```

```python
# Nenhuma célula nesse conjunto está vazia, isso não significa que todas tem dados, segundo a documentação as células vazias
# foram preenchidas com #NULO#
print((((df.iloc[:,:]=="#NULO#").sum())*100)/2000000)
```

```python
# As colunas restantes estão muito bem preenchidas, exceto a coluna DS_COMPOSICAO_FEDERACAO que apresentou 84% de células vazias,
# contudo isso não se configuira como uma falha afinal, segudo a documentação, um candidato pode ou não pertencer a uma federação
# como um número muito alto de células dessa coluna é composto de células vazias foi decidido removê-la
df.drop(['DS_COMPOSICAO_FEDERACAO'],axis=1,inplace=True)
```

```python
pd.set_option('display.width',320)
pd.set_option('display.max_columns',None)
plt.figure(figsize=(10,10))
sb.heatmap(df.corr())# Exibe um mapa de calor para a correlação de variáveis, aquelas que tiverem valores próximos de 1 ou -1
# estão muito relacionadas, por isso não há sentido em manter ambas
# aquelas que ficarem com correlação maior que |0.7| tem uma correlação forte
# entre |0.5| e |0.7| uma correlação moderada
# entre |0.3| e |0.5| uma correlação fraca
# entre 0 e |0.3| uma correlação despresível
plt.show()
```

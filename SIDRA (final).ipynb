```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

conjunto1 = pd.read_excel('C:/Users/caiom/IA/Arquivos/TCC/BD/SIDRA/LSPA - Área colhida, por período da safra e produto (hectares).xlsx')
conjunto2 = pd.read_excel('C:/Users/caiom/IA/Arquivos/TCC/BD/SIDRA/LSPA - Área plantada, por período da safra e produto (hectares).xlsx')
conjunto3 = pd.read_excel('C:/Users/caiom/IA/Arquivos/TCC/BD/SIDRA/LSPA - Produção, por período da safra e produto (toneladas).xlsx')
conjunto4 = pd.read_excel('C:/Users/caiom/IA/Arquivos/TCC/BD/SIDRA/LSPA - Rendimento médio, por período da safra e produto (quilogramas por hectare).xlsx')

conjunto1.drop([0,1,2], inplace = True)
conjunto1.rename(columns={'LSPA - Área colhida, por período da safra e produto (hectares)': 'Produto','Unnamed: 1':'Área colhida 2021 (hectares)','Unnamed: 2':'Área colhida 2022 (hectares)','Unnamed: 3': 'Variação área colhida'}, inplace = True)
conjunto1.dropna(inplace = True)

conjunto2.drop([0,1,2], inplace = True)
conjunto2.rename(columns={'LSPA - Área plantada, por período da safra e produto (hectares)': 'Produto','Unnamed: 1':'Área plantada 2021 (hectares)','Unnamed: 2':'Área plantada 2022 (hectares)','Unnamed: 3': 'Variação área plantada'}, inplace = True)
conjunto2.dropna(inplace = True)

conjunto3.drop([0,1,2], inplace = True)
conjunto3.rename(columns={'LSPA - Produção, por período da safra e produto (toneladas)': 'Produto','Unnamed: 1':'Produção 2021 (toneladas)','Unnamed: 2':'Produção 2022 (toneladas)','Unnamed: 3': 'Variação produção'}, inplace = True)
conjunto3.dropna(inplace = True)

conjunto4.drop([0,1,2], inplace = True)
conjunto4.rename(columns={'LSPA - Rendimento médio, por período da safra e produto (quilogramas por hectare)': 'Produto','Unnamed: 1':'Rendimento médio 2021 (quilogramas por hectare)','Unnamed: 2':'Rendimento médio 2022 (quilogramas por hectare)','Unnamed: 3': 'Variação rendimento médio'}, inplace = True)
conjunto4.dropna(inplace = True)
```

```python
aux1=pd.merge(conjunto1,conjunto2, how='outer', indicator=False)
aux2=pd.merge(conjunto3,conjunto4, how='outer', indicator=False)
df=pd.merge(aux1,aux2, how='outer', indicator=False)
df.head()
```

```python
df.dtypes
```

```python
df['Produto']=df['Produto'].astype(str) #converte o tipo de uma variável para outro
df['Área colhida 2021 (hectares)'] = df['Área colhida 2021 (hectares)'].astype(float)
df['Área colhida 2022 (hectares)']=df['Área colhida 2022 (hectares)'].astype(float)
df['Variação área colhida']=df['Variação área colhida'].astype(float)
df['Área plantada 2021 (hectares)']=df['Área plantada 2021 (hectares)'].astype(float)
df['Área plantada 2022 (hectares)']=df['Área plantada 2022 (hectares)'].astype(float)
df['Variação área plantada']=df['Variação área plantada'].astype(float)
df['Produção 2021 (toneladas)']=df['Produção 2021 (toneladas)'].astype(float)
df['Produção 2022 (toneladas)']=df['Produção 2022 (toneladas)'].astype(float)
df['Variação produção']=df['Variação produção'].astype(float)
df['Rendimento médio 2021 (quilogramas por hectare)']=df['Rendimento médio 2021 (quilogramas por hectare)'].astype(float)
df['Rendimento médio 2022 (quilogramas por hectare)']=df['Rendimento médio 2022 (quilogramas por hectare)'].astype(float)
df['Variação rendimento médio']=df['Variação rendimento médio'].astype(float)
df.dtypes
```

```python
pd.set_option('display.width',320)
pd.set_option('display.max_columns',None)
plt.figure(figsize=(10,10))
sb.heatmap(df.corr())
plt.show()
```

```python
df.drop(['Variação área colhida','Variação área plantada','Variação produção','Variação rendimento médio'],axis=1,inplace=True)
df.head()
```

```python
df.drop([0], inplace = True)
df['Rendimento médio 2021 (quilogramas por hectare)'].fillna(df['Rendimento médio 2021 (quilogramas por hectare)'].median(),inplace = True)
df['Rendimento médio 2022 (quilogramas por hectare)'].fillna(df['Rendimento médio 2022 (quilogramas por hectare)'].median(),inplace = True)
df.head()
```
